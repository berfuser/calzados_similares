Captura de imagenes



---




```python
destino_url='/content/drive/MyDrive/Poyecto_Final_EML/Img_Calzado/'
```

```python
#creo una lista con las imagenes para poder guardarlos Drive
lista_calzados=calzado['Imagen'].unique()
```

```python
sin_imagen=[]
for i in lista_calzados:
  try:
   url=i
   destino=destino_url+i.replace("https://res.cloudinary.com/dgaaedmn4/image/upload/######/","")
   urllib.request.urlretrieve(url,destino)
  except:
    sin_imagen.append(i)
  print(destino)
```
